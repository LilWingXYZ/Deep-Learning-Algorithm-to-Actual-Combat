```python
import os, shutil
```

```python
data_root = "tiny-imagenet-200"
data_type="train"
boxes = "_boxes.txt"
image_folder = "images"
```

```python
for dic_name in os.listdir(data_root+ '/'+data_type):
    image_dic = data_root+ '/' + data_type + '/'+ dic_name + '/'+ image_folder
    images_files = os.listdir(image_dic)
    i = 0
    for image_name in images_files:
        image_path = image_dic + '/' + image_name
#         print(image_path)
        move_to_path = data_root+ '/' + data_type + '/'+ dic_name + '/' + image_name
        shutil.move(image_path, move_to_path)
#         print(move_to_path)
    print(image_dic)
    boxes_file = data_root+ '/' + data_type + '/'+ dic_name + '/' + dic_name + boxes
    print(boxes_file)
    if os.path.exists(boxes_file):
        os.remove(boxes_file)
    shutil.rmtree(image_dic)
```
